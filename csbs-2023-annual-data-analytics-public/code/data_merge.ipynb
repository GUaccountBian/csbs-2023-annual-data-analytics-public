# Data Merging

```python
import pandas as pd
```

```python
# Read through all folders in the Call_report_Data
def read_year(year):
    path1 = f'../data/Call_Report_Data/FFIEC CDR Call Bulk Subset of Schedules {year}/FFIEC CDR Call Subset of Schedules {year}(1 of 2).txt'
    path2 = f'../data/Call_Report_Data/FFIEC CDR Call Bulk Subset of Schedules {year}/FFIEC CDR Call Subset of Schedules {year}(2 of 2).txt'
    df1 = pd.read_csv(path1,sep='\t')
    df2 = pd.read_csv(path2,sep='\t')
    df =  pd.concat([df1,df2],axis=1)
    return df
    
```

```python
df_2004 = read_year(2004)
df_2005 = read_year(2005)
df_2006 = read_year(2006)
df_2007 = read_year(2007)
df_2008 = read_year(2008)

df_2019 = read_year(2019)
df_2020 = read_year(2020)
df_2021 = read_year(2021)
df_2022 = read_year(2022)
```

```python
new_col = set(df_2004.columns).intersection(set(df_2006.columns),set(df_2005.columns),set(df_2007.columns),set(df_2008.columns),set(df_2019.columns),set(df_2020.columns),set(df_2021.columns),set(df_2022.columns))
```

```python
new_col
```

```python
new_col = list(new_col)
new_col = [i for i in new_col if len(i) < 10]
new_col = new_col+ ['Reporting Period End Date','IDRSSD','FDIC Certificate Number','OCC Charter Number','OTS Docket Number','Primary ABA Routing Number','Financial Institution Filing Type']
```

```python
df = pd.DataFrame(columns=new_col)
```

```python
def fil_df(df):
    df = df[new_col]
    return df
```

```python
df_2004 = fil_df(df_2004)
df_2005 = fil_df(df_2005)
df_2006 = fil_df(df_2006)
df_2007 = fil_df(df_2007)
df_2008 = fil_df(df_2008)

df_2019 = fil_df(df_2019)
df_2020 = fil_df(df_2020)
df_2021 = fil_df(df_2021)
df_2022 = fil_df(df_2022)
```

```python
pd.concat([df_2004,df_2005,df_2006,df_2007],ignore_index=True).to_csv('../data/cleaned/04-08.csv')
```

```python
pd.concat([df_2019,df_2020,df_2021,df_2022],ignore_index=True).to_csv('../data/cleaned/19-22.csv')
```
